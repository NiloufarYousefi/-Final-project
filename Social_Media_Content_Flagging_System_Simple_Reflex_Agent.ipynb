```python

###Social Media Content Flagging System Simple Reflex Agent



def ask_platform_choice():
    # Ask the user to select the social media platform where the content was found
    print("Which social media platform was this content posted on?")
    print("1. Twitter")
    print("2. Facebook")
    print("3. Instagram")
    print("4. TikTok")
    print("5. Other")

    try:
        platform_choice = int(input().strip())
        if platform_choice == 1:
            platform = "Twitter"
        elif platform_choice == 2:
            platform = "Facebook"
        elif platform_choice == 3:
            platform = "Instagram"
        elif platform_choice == 4:
            platform = "TikTok"
        elif platform_choice == 5:
            platform = "Other"
        else:
            print("Invalid choice. Please select a number between 1 and 5.")
            ask_platform_choice()
            return

        print(f"Content found on: {platform}")
        ask_initial_question(platform)

    except ValueError:
        print("Invalid input. Please enter a number between 1 and 5.")
        ask_platform_choice()

def ask_initial_question(platform):
    # Ask the user if the content is inappropriate
    print("Is this content inappropriate? (Yes/No)")
    response = input().strip().lower()

    if response == "yes":
        ask_flagging_reason(platform)
    elif response == "no":
        print("Thank you for your feedback. This content has not been flagged.")
    else:
        print("Invalid response. Please type 'Yes' or 'No'.")
        ask_initial_question(platform)

def ask_flagging_reason(platform):
    # Ask the user to select a reason for flagging
    print("Please select the reason for flagging this content:")
    print("1. Offensive Language")
    print("2. Hate Speech")
    print("3. Harassment")
    print("4. Violence")
    print("5. Misinformation")
    print("6. Other")

    try:
        reason_choice = int(input().strip())
        if reason_choice == 1:
            reason = "Offensive Language"
        elif reason_choice == 2:
            reason = "Hate Speech"
        elif reason_choice == 3:
            reason = "Harassment"
        elif reason_choice == 4:
            reason = "Violence"
        elif reason_choice == 5:
            reason = "Misinformation"
        elif reason_choice == 6:
            reason = "Other"
        else:
            print("Invalid choice. Please select a number between 1 and 6.")
            ask_flagging_reason(platform)
            return

        print(f"Thank you for reporting this content from {platform}. You flagged it for: {reason}.")
        print(f"Our team will review it soon on {platform}.")

        ask_additional_details()

    except ValueError:
        print("Invalid input. Please enter a number between 1 and 6.")
        ask_flagging_reason(platform)

def ask_additional_details():
    # Ask the user if they want to provide additional details (such as hashtags, usernames, etc.)
    print("Would you like to provide any additional details? (e.g., hashtags, usernames, etc.) (Yes/No)")
    response = input().strip().lower()

    if response == "yes":
        print("Please enter the details you want to provide (e.g., @username, #hashtag):")
        additional_details = input().strip()
        print(f"Additional details provided: {additional_details}")
    elif response == "no":
        print("No additional details provided.")
    else:
        print("Invalid response. Please type 'Yes' or 'No'.")
        ask_additional_details()

# Main function to start the process
def main():
    print("Welcome to the Social Media Content Flagging System!")
    ask_platform_choice()

# Run the program
if __name__ == "__main__":
    main()

```
