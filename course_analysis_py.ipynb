```python
# course-analysis.py (Google Drive version for Colab)

from google.colab import drive
import os

# Step 1: Mount Google Drive
drive.mount('/content/drive')

# Step 2: Set your Google Drive folder path (edit this if needed)
# Example: if your files are in "My Drive/Colab Notebooks/courses/"
folder_path = '/content/drive/My Drive'

# Step 3: Define function to extract course names
def get_courses(filename):
    courses = set()
    with open(filename, encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            if line.endswith('unit') or line.endswith('units'):
                try:
                    last_comma_index = line.rindex(',')
                    course_name = line[:last_comma_index].strip()
                    if course_name.endswith('*'):
                        course_name = course_name[:-1].strip()
                    courses.add(course_name)
                except ValueError:
                    continue
    return courses

# Step 4: Define main function
def main():
    file1 = os.path.join(folder_path, 'course-list1.txt')
    file2 = os.path.join(folder_path, 'course-list2.txt')
    file3 = os.path.join(folder_path, 'course-list3.txt')

    for f in [file1, file2, file3]:
        if not os.path.exists(f):
            print(f"❌ File not found: {f}")
            return

    courses1 = get_courses(file1)
    courses2 = get_courses(file2)
    courses3 = get_courses(file3)

    common_courses = courses1 & courses2 & courses3

    print("\n✅ Common Courses:")
    for course in sorted(common_courses):
        print(course)

# Step 5: Run the program
main()
```
